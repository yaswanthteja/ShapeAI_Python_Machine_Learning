### We convert our dataset into the pandas dataframe. so that it is easier to analyse the data

```python
boston = pd.DataFrame(df.data, columns=df.feature_names)
boston.head()
```

### Adding a new column of target values to the dataframe

```python
boston['MEDV'] = df.target
boston.head()
```

### Cheking datset contains any null"value or not

```python
boston.isnull()
```

### isnull retun True or False for each of the cell in the dataframe, but we can't go exploring all the cells to look for True values if any. so we use sum()function to count all the cells with True value (i.e. Null cells)

```python
boston.isnull().sum()
```

### We never train the model on all the data that we have, we always make sure to atleast have a test dataset, which is different from the training dataset.

```python
from sklearn.model_selection import train_test_split
X = boston.drop('MEDV', axis=1)
Y=boston['MEDV']
X_train, X_test, Y_train,Y_test = train_test_split(X,Y,test_size =0.15,random_state=5)
print(X_train.shape)
print( X_test.shape)
print(Y_train.shape)
print(Y_test.shape)
```

### Now let's import the Linear Regression model from sklearn and train it on the training dataset

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
#FITTING MODEL THE TRAINING DATASET

lin_model = LinearRegression()
lin_model.fit(X_train, Y_train)
```

```python
y_train_predict =lin_model.predict(X_train)
rmse = (np.sqrt(mean_squared_error(Y_train, y_train_predict)))
print ("The model performance for training set")
print( 'RMSE is {}'.format(rmse))
print("\n")

# on testing set

y_test_predict = lin_model.predict(X_test)
rmse = (np.sqrt(mean_squared_error(Y_test, y_test_predict)))

print("The model performance for testing set")
print('RMSE is {}'.format(rmse))
```
